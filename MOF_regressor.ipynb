```python
import csv
import numpy as np
from util.mof_util import test_regressor, tune_regressor
from util.import_data import import_data
from sklearn.ensemble import ExtraTreesRegressor, GradientBoostingRegressor, RandomForestRegressor
from matplotlib import pyplot as plt
```

```python
extra_tree = ExtraTreesRegressor()
test_regressor(extra_tree, data_x, bulk_y, shear_y, testsize=0.1, verbose=1)
```

```python
tune_regressor(extra_tree, data_x, bulk_y, shear_y, testsize=0.1, verbose=1)
```
